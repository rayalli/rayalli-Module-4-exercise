```python
# we are now fixing the students names 
# add each prefix and suffix to remove in a list remember to care about spaces!!!

prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]

# Iterate through the "prefixes_suffixes" list and replace them with an empty space, "" when it appears in the student's name.
for word in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word,"")
    
```

```python
# we will merge both student_data_df and school_data_df for calculations

#We'll merge school_data_df and student_data_dfon a shared column using the merge() method. 
#The column that these DataFrames have in common is school_name. 
#Inside the parentheses of the merge() method, we'll do the following:

# Combine the data into a single dataset.
school_data_complete_df = pd.merge(student_data_df,school_data_df,on=["school_name","school_name"])

#check the merge in the the new school_data_complete_df
school_data_complete_df.head()



```
